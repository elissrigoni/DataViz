**Pros and cons of the infographics**
**Isarithmic map**- also known as Contour map, isopleth map, isochrone map
Similar to chloroplet map, isarithmic map is used when data and subregional borders do not match, and so there is no relation between gradient and borders. Interpolation may reduce the precision, but in this type of map precision is not what we are looking for since the purpose is providing an overall distribution of data across a geographical area. It is essential to have a legend close to map and colors of the borders should not interfere with the colors displaying the data. It is possible also to add isolines to enhance gradients.

Pros of Isopleth Maps:
1. Isopleth are more scientific than other methods of showing distribution and effectively show the distribution and variations.
2. This is especially useful for climatic maps such as isobars, isotherms, isohyet etc. and is known as the main tool for the meteorologists.
3. It is very easy to determine the gradient (rate of change).
4. Isopleth lines are independent of political boundaries and best suited to show the natural pattern of distribution of an element.

Cons of Isopleth Maps:
1. Many isopleths are drawn by interpolation which sometimes lead to inaccuracy in the directions of isopleths and the pattern of distribution shown by those lines is distorted.
2. It is confusing when there is large list of points or dots.
3. It is time consuming when there is large number of dots or points.

**Radar chart** - also known as filled radar chart, star chart, spider diagram, parallel coordinates
It is appropriate only for certain type of data: only if circular arrangement is linked to the dataset it represents. It is very useful to detect particular spikes and for being arranged in small multiples, namely to compare many samples in just one shot. It is composed by some basic elements: 

- Center point: This is the core of a spider chart (at the center) from which different axes are drawn.
- Axis: Each axis represents a variable in a radar chart and is given a name and different values. A radar graph has at least 3 axes.
- Grids: When axes are linked in a spider chart, it divides the entire graph into different grids that help us represent information in a better way.
- Values: Once the graph is drawn, we represent various values on each axis and plot the chart for every entry by allocating distinctive colors.

Pros of Radar Charts:
1. It acts as an excellent tool to compare different entities easily.
2. Even if there are multiple variables to compare, we can do it with a spider graph.
3. They are simpler to understand by the audience instead of a column diagram.
4. We can distinguish different entities via color- coding.
5. Extremely useful in drawing comparisons on the basis of different parameters
 

Cons of Radar Charts:
1. If there are a lot of variables to compare, then even a radar chart can get cluttered.
2. Sometimes, users find it hard to come up with an ideal data set to represent different variables together.
3. They are not ideal for making trade-off decisions or comparing vastly distinctive variables.
4. At times, the radial distance can be hard to judge and provide on-point information.
