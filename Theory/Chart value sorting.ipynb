**Chart value sorting is implemented through the so called LATCH RULE** (five ways to sort data in a chart):
1. **Location**: sequencing content according to spatial order, it is meaningful only if offers the most logical sequence in content readability (e.g. Broadway installation along a street).
2. **Alphabet**: alphabetical sorting facilitates efficient lookup and reference for the reader. It requires user to rearrange items according to values that the data are displaying. If you do not want to imply any ranking starting from the value, the sorting according to the lexicographic order is the more neutral way. (e.g. in case of statistics associated to ER for each US state, to be impartial the best way of representing the states is the alphabetical one, the ranking among state can be easily done by the reader).
3. **Time**: time-based sorting helps comparing changes over time. Thanks to efficient infographic a reader can easily compare same phenomena in different times and for different locations. (e.g. interest in seasonality of patterns makes chronological sorting meaningful).
4. **Category**: categorical sorting implies a data organization in different categories, ordered by a ranking inherited by their values or unique to the subject (e.g., in football team: goalkeeper, defenders, midfielders, attackers)
5. **Hierarchy**: hierarchical sorting is defined by increasing/decreasing quantities, efficient perception of size/distribution/ranking. In this case data itself is driving the sorting (e.g. US gender pay gap between male and female, in this case the value itself is used to order the graph).
