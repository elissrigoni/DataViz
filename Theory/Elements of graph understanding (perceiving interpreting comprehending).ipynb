Data visualization is the representation and presentation of data to facilitate understanding (Andy Kirk).
Main ingredients are:
* *Data* is the key actor in the dataviz process and when we talk about raw data, we refer to tabular form. The problem with raw data is that they cannot be compared and contrasted effectively.
* *Data representation* is the visualization of data through the combination of marks (all basic elements of visualization: points, lines, areas..) and attributes (all properties of marks). 
* *Data presentation* concerns all other visible design decision beyond the representation: interactivity, annotations, colors. All elements that does not regard data.
Facilitating understanding should be the goal of the whole process. The understanding mental process  involves three steps, where each one is dependent on the previous one:

1. **PERCEIVING** answers the question: what does it show? Concerns what is the first impression when looking and simply being able to read a chart efficiently and decode the representation of the data (graphical elements: shapes/sizes/colors). Concerned pretty automatic questions. 
+ ANSWER THE QUESTIONS:
*where are the largest/middle-sized/smallest values? (Size)*
*what proportion of the total does that value hold?*
*how do these values compare in ranking terms?*
*to which other values does this have a connected relationship?* 

2. **INTERPRETATION** answers the question: what does it mean? It concerns art of converting the perception into meaning use pre-existing knowledge and context to frame the implication of the visualization. Interpretation pays attention to the concept, to the topic. 
+ ANSWER THE QUESTIONS:
*is that relationship meaningful or insignificant?* 
*is it good to be big or it is better to be small?*
*what does it mean to go up or to go down?*
*is the decline/increase of that category especially surprising?*
This step depends on viewer's pre-existing knowledge about the subject and his capacity to utilise that knowledge to frame the implications of what has been read. If the viewer do not possess that knowledge, it may be up to the creator to address that lackness. to achieve this, the visualizer can use captions, headlines, colours and other annotated devices.

3. **COMPREHENSION** answers the question: what does it mean to me? the act of reasoning about the consequences of interpretation what is the new information carried by the data visualization used to represent the subject. 
+ ANSWER THE QUESTIONS:
*why is this relevant? What is the main messages? to whom?*
*Have I learned? has it confirmed what I suspected or enlightened me with new knowledge?*
*has this impacted me emotionally or left me indifferent?*
*Does the new understanding I have acquired do push me to take action (make a decision or change my behaviour) or it just enhance my knowledge about the subject?*
*Does the graph does provide a path of interpretation?*
