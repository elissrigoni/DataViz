**Step 1: Load libraries**

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import umap
import umap.umap_ as umap


%matplotlib inline
sns.set(style='white', context='notebook', rc={'figure.figsize':(14,10)})
```

**Step 2: Importing the data set Import the dataset and distributing the dataset into X and y components for data analysis.**

```python
df = pd.read_csv("madelon.csv", sep = " ", header = None)
df = df.dropna()
df
```

```python
# define x 
x = df.iloc[: , :-1]
# define target variable: select last column
y = df.iloc[: , -1].values
```

**Step 3: Splitting the dataset into the Training set and Test set**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

**Step 4: Feature Scaling Doing the pre-processing part on training and testing set such as fitting the Standard scale.**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**Step 4: construct a UMAP object**

```python
import umap
import umap.umap_ as umap
reducer = umap.UMAP(n_components = 2, n_neighbors = 30, min_dist = 0.3, random_state = 42)
```

**Step 5: Now we need to train our reducer, letting it learn about the manifold. We will use the `fit_transform` method which first calls fit and then returns the transformed data as a numpy array.**

```python
embedding = reducer.fit_transform(X_train)
```

**Step 6: display the plot**

```python
sns.scatterplot(x = embedding[:, 0], 
                y = embedding[:, 1], s = 20, hue = y_train, 
                palette = sns.color_palette('tab20', 2))

plt.gca().set_aspect('equal', 'datalim')
plt.legend(bbox_to_anchor = (1, 1))
plt.title('UMAP projection', fontsize=24)
```

**FINALLY**

**We do the same for the test data.**

```python
from sklearn.linear_model import LogisticRegression 
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
# Predicting the test set result using
# predict function under LogisticRegression
y_pred = classifier.predict(X_test)
```

```python
# making confusion matrix between
#  test set of Y and predicted value.
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
